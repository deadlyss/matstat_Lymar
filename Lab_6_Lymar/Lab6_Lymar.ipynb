* Написати функцію `pop_n()`, що видаляє елементи стеку з його початку до номеру `n` включно.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = new_node
            self.rear = new_node
            return
        self.rear.next = new_node
        self.rear = new_node

    def dequeue(self):
        if self.front is None:
            return None
        temp = self.front
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return temp.data

    def print_n(self, n):
        current = self.front
        count = 0
        while current and count < n:
            print(current.data)
            current = current.next
            count += 1
```

* Оцінити асисптотичну складність (в середньому і в найгіршому випадку) процедур `search`, `insert` і `delete` роботи зі стеком.

Оцінка асимптотичної складності для операцій:

Search: Для пошуку конкретного елемента, потрібно перебирати всі елементи стеку, що займає $O(n)$ часу в найгіршому випадку.
Найгірший випадок: $O(n)$

Insert:
Вставка елемента в стек виконується за постійний час $O(1)$, оскільки новий елемент просто додається на вершину стеку. Середня складність: $O(1)$ Найгірший випадок: $O(1)$

Delete:
Видалення елемента зі стеку виконується за постійний час $O(1)$, оскільки видаляється тільки верхній елемент стеку. Середня складність: $O(1)$ Найгірший випадок: $O(1)$

## Контрольні запитання

__1.__ Що таке стек і які операції можна виконувати зі стеком?

Стек – це базова структура даних, в якій ми можемо працювати лише з елементами на початку стека, тобто вставляти або видаляти. Можна уявити як стопку тарілок, де можна додавати та вилучати лише верхні тарілки.

__2.__ Яка основна відмінність між стеком та чергою?

Основна відмінність між стеком і чергою полягає в тому, що у стеку ми можемо працювати лише з елементами на початку стеку, а у черзі ми додаємо елементи в один кінець, а видаляємо з іншого.

__3.__ Як ви можете реалізувати стек за допомогою масиву і за допомогою зв’язаного списку? Які переваги та недоліки кожного підходу?

Стек можна реалізувати за допомогою масиву, де верхній елемент стеку завжди знаходиться на останньому індексі масиву.

Переваги реалізації за допомогою масиву:

* є досить простою та інтуїтивно зрозумілою

* операції push і pop зазвичай виконуються за константний час O(1)

* легко ітерувати через всі елементи стеку

Стек також можна реалізувати за допомогою зв’язаного списку, де кожен елемент стеку є вузлом з посиланням на наступний елемент стеку.

Переваги реалізації за допомогою списку:

* список автоматично розширюється або скорочується при додаванні або вилученні елементів

* немає обмежень на розмір списку

__4.__ Які є застосування стека та черги в програмуванні і реальному житі?

У програмуванні: коли викликається функція в програмі, її контекст зберігається в стеці викликів. Це дозволяє програмі відслідковувати послідовність викликів функцій і правильно повертатися до попередніх контекстів після завершення функції.

Черги використовуються для управління обробкою завдань в системах, де потрібна обробка запитів або задач у порядку їхнього надходження.

У реальному житті: застосування черг у різних сценаріях обслуговування клієнтів черги допомагають організувати та управляти порядком обслуговування клієнтів в такому порядку, в якому вони надходять.
