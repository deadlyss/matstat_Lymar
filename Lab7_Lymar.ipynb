* Написати процедуру генерації купи з будь-якого рандомного масива

```python
import heapq
import random

# Створюємо випадковий масив з 10 елементів
random_array = [random.randint(1, 50) for _ in range(10)]

# Виводимо випадковий масив
print("Випадковий масив:", random_array)

# Перетворюємо масив на купу
heapq.heapify(random_array)

# Виводимо масив, перетворений на купу
print("Купа:", random_array)
```

* Додати до нього елемент, який дорівнює вашому порядковому номеру у списку групи

```python
# Додаємо елемент "11" до купи
heapq.heappush(random_array, 11)

# Виводимо оновлену купу
random_array
```

* Вилучити максимальний елемент з купи

```python
# Знайти максимальний елемент у купі
max_element = max(random_array)

# Видалити максимальний елемент з купи
random_array.remove(max_element)

# Перетворити масив назад на купу, щоб зберегти властивості купи
heapq.heapify(random_array)

# Вивести максимальний елемент і оновлену купу
print("Максимальний елемент:", max_element)
print("Оновлена купа:", random_array)
```

* Оцінити асимптотичну складність (в середньому і в найгіршому випадку) процедур `search`, `insert` і `delete` роботи з з купою.

__Пошук__ у купі займає лінійний час, оскільки купа не є впорядкованою структурою. У гіршому випадку потрібно перевірити всі $n$ елементів, щоб знайти необхідний. Тому у середньому та найгіршому випадку - $O(n)$

__Вставка__ нового елемента у купу вимагає додавання елемента в кінець масиву (або дерева) і підняття цього елемента до відповідного місця. Цей процес включає підйом через рівні дерева, що в середньому і в найгіршому випадку займає $O(log n)$.

__Видалення__ кореневого елемента потребує заміни цього елемента останнім елементом у масиві, а потім відновлення властивостей купи шляхом опускання нового кореня до відповідного місця. Цей процес включає проходження по рівнях дерева, що також займає $O(log n)$ у середньому та найгіршому випадках.

## Контрольні запитання

1. Чим відрізняється структура _бінарне дерево_ від _бінарного дерева пошуку_?


Бінарне дерево: може бути довільним деревом, що означає, що немає вимог до впорядкування значень вузлів. Всі рівні дерева можуть бути заповнені нерівномірно.

Бінарне дерево пошуку: для кожного вузла всі значення в лівому піддереві менші за значення вузла. Всі значення в правому піддереві більші за значення вузла, час пошуку, вставки і видалення елементів в середньому складає O(log n).

2. Чим відрізняється структура _бінарне дерево_ від _бінарної купи_?

Бінарне дерево та бінарна купа мають різні властивості та призначення. Бінарне дерево це деревоподібна структура даних, в якій кожен вузол має не більше двох дочірніх вузлів

Бінарна купа це повне бінарне дерево, що задавольняє властивість купи. Властивість купи: гарантує що найм. або найб. елемент знаходяться в корені дерева.

Бінарне дерево пошуку оптимальне для швидкого пошуку, тоді як бінарна купа корисна для швидкого доступу до мінімальних або максимальних елементів.
